```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("Data/merged_data.csv")

# For demonstration, let's predict 'Pesticide' using the other columns
X = df.drop(['Pesticide', 'Country Name', 'Country Code'], axis=1)
y = df['Pesticide']

# If there are categorical columns, convert them
X = pd.get_dummies(X)

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Fit the decision tree regressor
tree = DecisionTreeRegressor(max_depth=8, random_state=42)
tree.fit(X_train, y_train)

# Plot the tree
plt.figure(figsize=(20, 10))
plot_tree(tree, feature_names=X.columns, filled=True)
plt.show()


```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

ypred = tree.predict(X_test)

# Calculate regression metrics
mse = mean_squared_error(y_test, ypred)
mae = mean_absolute_error(y_test, ypred)
r2 = r2_score(y_test, ypred)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R^2 Score:", r2)

```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Load the data
df = pd.read_csv("Data/merged_data.csv")

df = df[df['Pesticide'] <= 200000]
df = df[df['Ag_land_Area'] >= 2000]
# Prepare features and target
X = df.drop(['Pesticide', 'Country Name', 'Country Code'], axis=1)
y = df['Pesticide']

# Convert categorical columns if any
X = pd.get_dummies(X)

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.11, random_state=42)

# Fit the Random Forest regressor
forest = RandomForestRegressor(n_estimators=1000, max_depth=6, random_state=42)
forest.fit(X_train, y_train)

# Predict and evaluate
ypred = forest.predict(X_test)
mse = mean_squared_error(y_test, ypred)
mae = mean_absolute_error(y_test, ypred)
r2 = r2_score(y_test, ypred)

print("Random Forest Results:")
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R^2 Score:", r2)
```
