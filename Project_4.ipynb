```python
land_area_path = "Data/Pesticide.csv"


df = pd.read_csv(land_area_path)


df_p = df[df['COMP_BREAKDOWN_1_LABEL'].str.contains('total', case=False, na=False)]

# Read the CSV, skipping the first 4 lines and making the next row the header
##df = pd.read_csv(land_area_path, skiprows=4)

# Rename '2020' column to 'Pesticide'
df_pp = df_p.rename(columns={"2020": "Pesticide"})

# Keep rows where the 'Item: Pesticides (total)' column contains 'total'
#df_total = df[df['Item: Pesticides (total)'].str.contains('total', case=False, na=False)]
df_pp = df_pp.rename(columns={"REF_AREA": "Country Code"})
# Keep only the 'REF_AREA' (country code) and 'Pesticide' columns
df_pesticide = df_pp[["Country Code", "Pesticide"]]

df_pesticide

```

```python
# Load the Ag_Land_Area CSV and keep only 'Country Code' and 'Country Name' columns
df_key = pd.read_csv("Data/Ag_Land_Area.csv", skiprows=4)
df_key = df_key[["Country Code", "Country Name"]]
df_key
```

```python
from functools import reduce

dfs = [df_AG_AREA, df_AG_pct, df_pct_gdp, df_cereal_yield, df_GDP, df_pesticide, df_key]
df_merged = reduce(lambda left, right: pd.merge(left, right, on="Country Code", how="outer"), dfs)

# Display all rows in the dataframe
import pandas as pd
pd.set_option('display.max_rows', 8)

display(df_merged)
```

```python
# Drop all rows with NaNn values in the 'Pesticide' column

df_merged = df_merged.dropna(subset=['Pesticide', 'Country Name'])

pd.set_option('display.max_rows', None)
display(df_merged)

### I have the datafram I am going to work with now. 
```

```python
# Drop rows that have NaN values in any column
df_merged = df_merged.dropna()
display(df_merged)

# Convert scientific notation to numbers for display
import pandas as pd
pd.set_option('display.float_format', '{:.2f}'.format)


df_merged.to_csv("Data/merged_data.csv", index=False)
```
